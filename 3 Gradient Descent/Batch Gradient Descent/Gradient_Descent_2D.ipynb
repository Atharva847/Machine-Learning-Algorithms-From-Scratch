```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('package.csv')
```

```python
X = df['CGPA'].values
y = df['Package_LPA'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state = 5, test_size = 0.2)
```

```python
class GradientDescent:

    def __init__(self, epochs, lr):

        self.epochs = epochs
        self.lr = lr
        self.m = 0
        self.b = 0

    def fit(self, X, Y):

        n = len(X)

        for i in range(self.epochs):

            y_hat = self.m * X + self.b
            error = Y - y_hat

            loss_slope_m = -(1/n) * np.sum(error * X)
            loss_slope_b = -(1/n) * np.sum(error)

            self.m -= self.lr * loss_slope_m
            self.b -= self.lr * loss_slope_b

        print(f'slope : {self.m}, intercept : {self.b}')

    def predict(self, X):

        return self.m * X + self.b
```

```python
gdr = GradientDescent(100000, 0.01)
```

```python
gdr.fit(X_train, Y_train)
```

```python
y_pred = gdr.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color='blue', label='Train Data',alpha=0.6, s=50)
plt.scatter(X_test, Y_test, color='green', label='Test Data', alpha=0.6, s=50)
plt.plot(X_test, y_pred, color = 'red', label = 'Line')
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print("R2 Score is : ", r2_score(Y_test, y_pred))
```

```python
print('MSE is : ', mean_squared_error(Y_test, y_pred))
```

```python
print('MAE is : ', mean_absolute_error(Y_test, y_pred))
```
